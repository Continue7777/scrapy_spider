```python
import pandas as pd 
```

```python
df = pd.read_csv("result_detail.csv",sep=',',names=['platform','domain','owner_name','link','pixelId','facebook_url','like_num','share_num','comment_num'])
```

```python
domain_df = pd.read_csv("result/domain_detail.csv",sep='\t',names=['domain','flag','emails'])
```

```python
df = df.drop_duplicates()
df.tail()
```

```python
sum_df = df.groupby("domain")['like_num','share_num'].sum().reset_index()
```

```python
count_df = df.groupby("domain")['like_num'].count().reset_index()
count_df.columns = ['domain','count_num']
```

```python
final_df = df.drop_duplicates(["domain","owner_name"])
final_df = final_df[['platform','domain','owner_name','facebook_url']]
```

```python
final_df = pd.merge(pd.merge(pd.merge(final_df,sum_df),count_df),domain_df)
```

```python
final_df
```

```python
final_df.to_csv("result/result_gather.csv",index=False,sep='\t')
```
