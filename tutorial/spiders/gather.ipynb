```python
df = pd.read_csv("result/product_detail.csv",sep=',',names=['platform','owner_name','domain','link','pixelId','facebook_url','like_num','share_num','comment_num'])

df['like_num'] = df['like_num'].apply(lambda x: x if re.search("^\d+$", str(x)) else np.nan)
df['share_num'] = df['share_num'].apply(lambda x: x if re.search("^\d+$", str(x)) else np.nan)
df['comment_num'] = df['comment_num'].apply(lambda x: x if re.search("^\d+$", str(x)) else np.nan)
df = df.fillna(0)
df[['like_num', 'share_num','comment_num']] = df[['like_num', 'share_num','comment_num']].astype(int)
```

```python
domain_check_df = pd.read_csv('result/domain_check.csv',sep='\t',header=None,names=['domain','code'])
```

```python
domain_check_df_200 = domain_check_df[domain_check_df['code']=='200']
```

```python
df = df.drop_duplicates()
df.tail()
```

```python
sum_df = df.groupby("domain")['like_num','share_num'].sum().reset_index()
```

```python
count_df = df.groupby("domain")['like_num'].count().reset_index()
count_df.columns = ['domain','count_num']
```

```python
final_df = df.drop_duplicates(["domain","owner_name"])
final_df = final_df[['platform','domain']]
```

```python
final_df = pd.merge(pd.merge(pd.merge(final_df,sum_df),count_df),domain_check_df_200)
```

```python
final_df = final_df.drop_duplicates()
```
